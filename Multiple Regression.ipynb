```python
#Step1: Import libraries
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
#Step2: Load data on collab
from google.colab import files
uploaded = files.upload()
```

```python
#step3: read dataframe with pandas
df = pd.read_csv('3_houseprices.csv')
df
```

```python
#step4: correcting NaN value by taking the medium of other rows with code df.bedrooms.median()
#if you want to convert to integer whole number, use below code
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
#step5: fill the median number into NaN value
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
#fit model is used to train the model
# df[['area', 'bedrooms', 'age']]: create dataframe from existing dataframe with independent variables area, bedrooms, age
# df.price = target variable
reg= linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
