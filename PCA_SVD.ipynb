```python
df = pd.read_csv('Population and GDP by Country.csv', index_col='Country Name')
df.head()
```

```python
df = df.apply(pd.to_numeric)
```

```python
X=df.to_numpy()
X
```

```python
# calculated mean vector  
X_centered= X-X.mean(axis=0)
print(X_centered)
```

```python
# Perform SVD
U, s, Vt = np.linalg.svd(X_centered)

print(f' shape of U = {U.shape}')
print(f' shape of s = {s.shape}')
print(f' shape of Vt = {Vt.shape}')

# create a cov matrix S from s
m, n = X.shape
S = np.zeros(X_centered.shape)
S[:n, :n] = np.diag(s)

print(f' shape of S = {S.shape}')
```

```python
# for confirmation

u1 = U[:, 0]
u2 = U[:, 1]
u3 = U[:, 2]
u4 = U[:, 3]

v1 = Vt.T[:, 0]
v2 = Vt.T[:, 1]
v3 = Vt.T[:, 2]
v4 = Vt.T[:, 3]

print(f'X_centered = {X_centered.shape}')
print('\n=======')
print(f'U.shape = {U.shape}')
print(f's.shape = {s.shape}')
print(f'Vt.shape = {Vt.shape}')

print('\n=======')

print('left singular matrix U')
print('first (u1) to fourth vector (u4) of U')
print(f'u1  = {u1}')
print(f'u2  = {u2}')
print(f'u3  = {u3}')
print(f'u4  = {u4}')

print('\n=======')
print(f's = {s}' )
print(f'S (first 10 rows) = \n {S[:10]}')

print('\n=======')
print('right singular matrix V')
print('first (v1) to fourth vector (v4); they are in horizontal direction')
print(f'v1  = {v1}')
print(f'v2  = {v2}')
print(f'v3  = {v3}')
print(f'v4  = {v4}')
```

Perform PCA on X

```python
from sklearn.decomposition import PCA
pca4 = PCA(n_components=4)
X4D = pca4.fit_transform(X_centered)
COV_pca = pca4.get_covariance()
print(f' variance-covariance matrix calculated by sklearn :\n {COV_pca}')
print('\n this is the same with the above')
```

Calculation of unitary matrix Upca where each column is eigenvector of Xc (by PCA)

```python
# eigen vectors; u1, u2, u3, u4
U_pca=pca4.components_
print(f'eigenvector U_pca (in horizontal direction; U^T)= \n {U_pca}')

# as for comparison 
print('\n right singular matrix V by SVD')
print('first (v1) to fourth vector (v4); they are in horizontal direction')
print(f'v1  = {v1}')
print(f'v2  = {v2}')
print(f'v3  = {v3}')
print(f'v4  = {v4}')
```

Confirm they are the same 

```python
# eigenvalues - you will find the same.
eigenvalues = pca4.explained_variance_
print(f'PCA; eigenvalues of cov(X) = {eigenvalues}')

print(f'SVD: singular values of X = {s}')


print(f'\n  (singular value)^2 / (N-1)= \n  {s*s/(m-1)}')
```
